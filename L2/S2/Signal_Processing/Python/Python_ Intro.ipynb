###Advice about functions arguments
For your own functions, I advise you not to use optional arguments at all; which goes well with our motto "program as simply as possible".

But you will have to use functions created by others that have optional arguments. For example, consider a function of the scipy module (which will be detailed later). It generates Gaussian random values.

````
import scipy.stats as stats
""" 'loc':expected value, scale:'standard deviation', 'size': number of random values generated"""
X=stats.norm.rvs(loc=-1,scale=3,size=3)
print(X)
````

In the previous function, all arguments are optional. The default values are logically `loc = 0, scale = 1, size = 1`. We can write for example

     simus = stat.norm.rvs (size = 1000)
for

     simus = stat.norm.rvs (loc = 0, scale = 1, size = 1000)


But beware: if we do not specify the name of the arguments, they are taken in the order `1: loc 2: scale 3: size`
For example: if, to shoot 1000 Gaussian, I write

     simus = stat.norm.rvs (1000)

our program bug because that corresponds to

     simus = stat.norm.rvs (loc = 1000)

I advise you to write almost all the time the wname of the arguments to avoid this kind of confusion; except when it is an obvious mandatory argument as in `plt.plot (x, y)`.

###Common points
The lists, the tuples, the strings of character, the iterables (and the vectors numpy, cf further) all have in common to be traversed (= iterated). Many ready-made functions accept any of these types. This flexibility is part of the python philosophy.

To illustrate this, let us write a function that accepts many different arguments:

````
def sequence_to_string(sequence): 
    res = "" 
    for  i  in  range(len(sequence)): 
        res += str(sequence[i]) 
        if (i<len(sequence)-1): res +="_" 
    print(res)

sequence_to_string ([1,2]) #from list
sequence_to_string ((3,4)) #from tuple 
sequence_to_string ("foo") #from string 
sequence_to_string (range(3)) #from iterable 
sequence_to_string (np.arange(0,5)) #from numpy array
````

## Dictonaries Khaliq Aghakarimov

Dictionaries (= hash tables) are essential tools for programming. A dictionary is a collection containing VALUES that are indexed by unique KEYS. Often the keys are strings of characters.

The dictionaries are very efficient because the access time to the values (via a key) does not depend on the size of the dictionary. Imagine that each key is transformed, via a hash, into a memory address that points to a value.

There are dictionaries everywhere:

* your operating system creates dictionaries with as key all the words present in all the documents of your computer, and as value the address of the documents in question.
* search engines create dictionaries to index all keywords in all languages.
* databases create indexes for faster access to data.

### Fill a dico

````
""" creation of a dico by the notation 'JSON' (used by many languages)"""
dictionary = {'anInteger': 3, 'aString': 'aba'}
"""creation of a list by the JSON notation"""
myList = [1, 2, 3, 4, 5]

""" create a dico element by element """
dico = {}
dico['anInteger'] = 3
dico['aString'] = 'aba'
dico['aDico'] = dictionary
dico['aList'] = myList

dico
````

###Access to the elements

There are two ways to access values:

````
print ( dico ['anInteger'])   # when we are sure that the key is present. 
print ( dictionary.get('foo'))   # when it is not safe. This returns None when the key is not present
````

Each key in a dictionary is unique. Dictionaries can be used to represent lists without repetition:

````
aList = [1, 2, 1, 4, 1, 4, 1, 1, 1, 1]
aListWithoutRepetition = {}
for i in aList:
    aListWithoutRepetition[i] = True
print(aListWithoutRepetition)
print("0 is in our set ?:", aListWithoutRepetition.get(0))
````

Dictionaries are often used to count occurrences.

````
aPhrase = "hello hello to you how are you going"
theWords = aPhrase.split(sep=" ")
print(theWords)
dico = {}
for word in theWords:
    if dico.get(word) is None:
        dico[word] = 1
    else:
        dico[word] += 1
        
dico
````

````
"""a trick to go faster"""
dico2 = {}
for word in theWords:
    """dico.get (key, 0) returns 0 (instead of None) when the key is not present"""
    dico2[word]=dico.get(word,0)+1
dico2
````

### Application: find common elements

Suppose we want to create a function that quickly calculates the elements common to two lists. For example with as input:

```
    list1=[1,3,5,7]
    list2=[8,6,5,3,2]
```

this function will return `[3,5]`or `[5,3]`. But let's take longer lists so that the execution time is not negligible:



````
list1=range(0,300000,33)
list2=range(0,300000,17)
````

Naively, to calculate this intersection, we would do a double loop, but it's very slow:

````
%%time
inter=[]
for i in list1:
    for j in list2:
        if i==j: inter.append(j)
print(len(inter))
````

The good technique is to transform the first list in dico/set then to loop on the second list:

````
%%time
list1_as_set=set(list1)
inter=[]
for elem in list2:
    if elem in list1_as_set:
        inter.append(elem)
        
print(len(inter))
````

### Create your own object

The objects are very useful for organizing the code. It is a computer structure that brings together:

* Data, ex: self.nameand self.age (see below)
* Methods, ex: self.make_it_older(self).

````
class Man:
    
    "this function is started automatically when we create the object"
    def __init__(self,name,age):
        self.name=name
        self.age=age
        
    "the def here are methods = functions attached to the object"
    def make_it_older(self):
        self.age+=1
        
    def change_name(self,new_name):
        self.name=new_name
        
    def print_me(self):
        print(self.name,", age:",self.age)
        
        
toto=Man("toto",7)
toto.print_me()
toto.make_it_older()
toto.print_me()
````

The instruction `toto.print_me()`can also be written  `Man.print_me(toto)`. That's why there is the word `self` in all methods: do not forget it.
