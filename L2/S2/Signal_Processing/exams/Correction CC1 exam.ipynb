* $(1\heartsuit)$ What is the discrete derivative of a vector?

Take a vector $u$.  Its discrete derivative is $u'[n]=u[n]-u[n-1]$

---

* $(1\heartsuit)$ How can we do this  with a convolution?  

We take a mask $m=[1,-1]$ and we perform $u'=u\star m$. 

---

* $(1\heartsuit)$ What is the second discrete derivative of a vector? 

Take a vector $u$.  Its discrete second derivative is 
$$
u''[n]=u'[n]-u'[n-1]=u[n]-2u[n-1]+u[n-2]
$$

---

* $(1\heartsuit)$ How can we do this  with a convolution?

We take a mask $m=[1,-2,1]$ and we perform $u''=u\star m$. 


---

* $(1\heartsuit)$ How can we smooth a vector with a convolution?
