## Definition Khaliq Aghakarimov

A realy usefull basis of vectors is the Discrete-Cosinus-basis $(b^0,b^1,...,b^{N-1})$ defined by:
$$
b^k_j = c_k \cos\Big( (2 j +1)  \frac{ \pi k}  {2N}  \Big)
$$
with  $c_0 = \sqrt{\frac 1 N}$ and for $k>0: $ $ c_k = \sqrt{\frac 2 N}$.

```python
import numpy as np
import matplotlib.pyplot as plt

np.set_printoptions(linewidth=5000,suppress=True,precision=5)
plt.style.use("default")
```

```python
N=8
basis=np.empty([N,N])
i=np.arange(N)
for k in range(N):
  basis[k,:]=np.cos((2.*i+1)*np.pi*k/(2.*N))
  if k==0:
    basis[k,:]*=np.sqrt(1/N)
  else:
    basis[k,:]*=np.sqrt(2/N)
```

```python
fig,axs=plt.subplots(N,1,figsize=(6,N),sharex=True)
for k in range(N):
  axs[k].plot(i,basis[k,:],'.')
  axs[k].set_title("$b^k_j = c_k \cos (2 j +1)   \frac{ \pi k} {2N} )$")
  axs[k].set_ylim(-1,1)
fig.tight_layout()
```

 Are they orthonormals ?

```python
all_scalar_products=basis@basis.T # check whether we get Identity matrix
print(all_scalar_products)
```
